```python
# Defining the cost of  connected nodes as a 2x2 matrix

COSTS = [
    
        [0, 7, 9,'inf','inf',14],
        [7, 0, 10, 15, 'inf','inf'],
        [9, 10, 0, 11, 'inf', 2],
        ['inf', 15, 11, 0, 6, 'inf'],
        ['inf','inf','inf',6,0,9],
        [14, 'inf', 2, 'inf', 9,0]
    
]
```

```python
# Importing the Libraries to define the search problem and apply uniform cost search algorithm
from simpleai.search import SearchProblem, uniform_cost 
```

```python
# Problem Formulation, we need to define initial state, goal state, actions, results, goal_test and path cost from current state to next state

class Route(SearchProblem):
    
    def __init__(self, initial,goal):
        
        # The route start from 1 to 6 but in python index starts from 0 so 
        self.initial = initial-1
        self.goal = goal-1
        
        super(Route,self).__init__(initial_state=self.initial)
    
    # Defining actions
    def actions(self,state):
        actions = []
        
        for action in range(len(COSTS[state])):
            if COSTS[state][action] not in ['inf',0]:
                actions.append(action)
        return actions
    
    # the result state is just action as defined
    def result(self, state, action):
        return action
    
    # check if the goal is reached
    def is_goal(self, state):
        return state == self.goal
    
    # return the cost between the two nodes
    def cost(self, state, action, state2):
        return COSTS[state][action]
```

```python
# Once the problem is defined we can use the search algorithm
problem = Route(1,5)

result = uniform_cost(problem)

path = [x[1]+1 for x in result.path()]

print(f"the route is {path} and total cost is {result.cost}")
```
